```python
# Create a copy of the original dataset

df_processed = df.copy()

print(df_processed.shape)

df_processed.head()
```

```python
# Binary Classification Target

y_failure = df_processed["Machine failure"]

# Multi-label Classification Target

y_failure_type = df_processed[
    ["TWF", "HDF", "PWF", "OSF", "RNF"]
]
```

```python
X = df_processed.drop(
    columns=[
        "Machine failure",
        "TWF",
        "HDF",
        "PWF",
        "OSF",
        "RNF"
    ]
)

print(X.columns)
```

```python
X.drop(
    columns=[
        "UDI",
        "Product ID"
    ],
    inplace=True,
    errors='ignore'
)
```

```python
X = pd.get_dummies(
    X,
    columns=["Type"],
    drop_first=True
)
```

```python
print("Feature Shape :", X.shape)

print("Machine Failure Shape :", y_failure.shape)

print("Failure Type Shape :", y_failure_type.shape)

X.head()
```

```python
from sklearn.model_selection import train_test_split

(
X_train,
X_test,
y_failure_train,
y_failure_test,
y_type_train,
y_type_test
)=train_test_split(

X,

y_failure,

y_failure_type,

test_size=0.20,

random_state=42,

stratify=y_failure

)
```

```python
print("Training Set Distribution")
print(y_failure_train.value_counts())

print("\nPercentage Distribution")
print(y_failure_train.value_counts(normalize=True) * 100)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(
    random_state=42,
    max_iter=1000
)

log_model.fit(
    X_train,
    y_failure_train
)
```

```python
y_pred = log_model.predict(X_test)

y_prob = log_model.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
print("Accuracy :", accuracy_score(y_failure_test, y_pred))

print("Precision :", precision_score(y_failure_test, y_pred))

print("Recall :", recall_score(y_failure_test, y_pred))

print("F1 Score :", f1_score(y_failure_test, y_pred))

print("ROC AUC :", roc_auc_score(y_failure_test, y_prob))
```

```python
from sklearn.metrics import classification_report

print(classification_report(
    y_failure_test,
    y_pred
))
```

```python
from sklearn.metrics import confusion_matrix

import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(
    y_failure_test,
    y_pred
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['No Failure','Failure'],
    yticklabels=['No Failure','Failure']
)

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.title("Confusion Matrix - Logistic Regression")

plt.show()
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(
    y_failure_test,
    y_prob
)

plt.figure(figsize=(7,5))

plt.plot(fpr, tpr, label="Logistic Regression")

plt.plot([0,1],[0,1],'--')

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()
```

```python
baseline_results = {
    "Model":"Logistic Regression",
    "Accuracy":accuracy_score(y_failure_test,y_pred),
    "Precision":precision_score(y_failure_test,y_pred),
    "Recall":recall_score(y_failure_test,y_pred),
    "F1":f1_score(y_failure_test,y_pred),
    "ROC_AUC":roc_auc_score(y_failure_test,y_prob)
}

baseline_results
```

Engineering Interpretation

This baseline reveals two important characteristics of the dataset:

The operational variables contain enough information to separate healthy and failed machines, as indicated by the high ROC-AUC.
Severe class imbalance causes the classifier to favor the majority class, resulting in poor detection of machine failures.

Therefore, additional techniques such as class balancing and more powerful classifiers are expected to improve Recall significantly.

Applying class weighting produced no meaningful improvement in predictive performance. Instead, the balanced model showed a small decrease in Recall and F1-score while maintaining similar Accuracy, Precision, and ROC-AUC.

This suggests that the original Random Forest was already capable of handling the class imbalance effectively, likely because:

The engineered features provide strong separation between classes.
The imbalance (≈3.4% failures) is not extreme enough to require additional weighting.
Decision-tree ensembles are naturally more robust to moderate imbalance than linear models.


```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(
    random_state=42
)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_failure_train
)

print("Original Training Shape")
print(X_train.shape)

print("\nAfter SMOTE")
print(X_train_smote.shape)

print("\nClass Distribution After SMOTE")
print(y_train_smote.value_counts())
```

```python
rf_smote = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_smote.fit(
    X_train_smote,
    y_train_smote
)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report
)

print("Accuracy :", accuracy_score(y_failure_test, y_pred_smote))
print("Precision :", precision_score(y_failure_test, y_pred_smote))
print("Recall :", recall_score(y_failure_test, y_pred_smote))
print("F1 Score :", f1_score(y_failure_test, y_pred_smote))
print("ROC AUC :", roc_auc_score(y_failure_test, y_prob_smote))

print("\nClassification Report\n")
print(classification_report(y_failure_test, y_pred_smote))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(
    y_failure_test,
    y_pred_smote
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Purples",
    xticklabels=["No Failure", "Failure"],
    yticklabels=["No Failure", "Failure"]
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest + SMOTE")

plt.show()
```

```python
rf_smote_results = {
    "Model": "Random Forest + SMOTE",
    "Accuracy": accuracy_score(y_failure_test, y_pred_smote),
    "Precision": precision_score(y_failure_test, y_pred_smote),
    "Recall": recall_score(y_failure_test, y_pred_smote),
    "F1": f1_score(y_failure_test, y_pred_smote),
    "ROC_AUC": roc_auc_score(y_failure_test, y_prob_smote)
}

rf_smote_results
```

SMOTE successfully balanced the training data by generating synthetic failure samples. However, the oversampled data made the classifier more likely to predict the minority class, increasing false alarms without improving the detection of actual failures.

For predictive maintenance, excessive false positives may lead to:

unnecessary inspections,
increased maintenance costs,
reduced trust in the prediction system.

Therefore, despite its slightly higher ROC-AUC, the SMOTE-based model does not provide the best trade-off between Precision and Recall.

#Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
param_grid = {

    'n_estimators': [100, 200, 300, 500],

    'max_depth': [None, 10, 20, 30, 40],

    'min_samples_split': [2, 5, 10],

    'min_samples_leaf': [1, 2, 4],

    'max_features': ['sqrt', 'log2'],

    'bootstrap': [True, False]

}
```

```python
rf = RandomForestClassifier(
    random_state=42
)
```

```python
random_search = RandomizedSearchCV(

    estimator=rf,

    param_distributions=param_grid,

    n_iter=20,

    cv=5,

    verbose=2,

    random_state=42,

    scoring='f1',

    n_jobs=-1

)

# THIS IS REQUIRED
random_search.fit(X_train, y_failure_train)

# THEN
print(random_search.best_params_)
```

```python
print("Best CV F1 Score:")

print(random_search.best_score_)
```

```python
best_rf = random_search.best_estimator_

best_rf.fit(
    X_train,
    y_failure_train
)
```

```python
y_pred_best = best_rf.predict(X_test)

y_prob_best = best_rf.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report
)

print("Accuracy :", accuracy_score(y_failure_test, y_pred_best))

print("Precision :", precision_score(y_failure_test, y_pred_best))

print("Recall :", recall_score(y_failure_test, y_pred_best))

print("F1 Score :", f1_score(y_failure_test, y_pred_best))

print("ROC AUC :", roc_auc_score(y_failure_test, y_prob_best))

print(classification_report(
    y_failure_test,
    y_pred_best
))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(
    y_failure_test,
    y_pred_best
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['No Failure','Failure'],
    yticklabels=['No Failure','Failure']
)

plt.title("Optimized Random Forest")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python
best_rf_results = {

"Model":"Optimized Random Forest",

"Accuracy":accuracy_score(y_failure_test,y_pred_best),

"Precision":precision_score(y_failure_test,y_pred_best),

"Recall":recall_score(y_failure_test,y_pred_best),

"F1":f1_score(y_failure_test,y_pred_best),

"ROC_AUC":roc_auc_score(y_failure_test,y_prob_best)

}

best_rf_results
```

Hyperparameter tuning did not significantly improve the classification metrics because the baseline Random Forest was already well suited to the AI4I dataset. However, the increase in ROC-AUC demonstrates a modest improvement in the model's ability to distinguish between healthy and failed machines across different decision thresholds.

Therefore, the Optimized Random Forest is selected as the final binary classifier for machine failure prediction.

```python
def predict_complete_machine(sample):

    # -------- Machine Failure --------

    failure_prob = best_rf.predict_proba(sample)[0][1]

    failure_prediction = best_rf.predict(sample)[0]

    if failure_prediction == 0:

        return {

            "Machine Failure":"NO",

            "Confidence":failure_prob,

            "Failure Type":"None",

            "Recommendation":"Machine operating normally."

        }

    # -------- Failure Type --------

    probs = failure_model.predict_proba(sample)

    prob_dict = {}

    for col, p in zip(failure_columns, probs):

        prob_dict[col] = p[0][1]

    predicted = max(
        prob_dict,
        key=prob_dict.get
    )

    predicted_name = failure_names[predicted]

    return {

        "Machine Failure":"YES",

        "Confidence":failure_prob,

        "Failure Type":predicted_name,

        "Failure Confidence":prob_dict[predicted],

        "Recommendation":recommendations[predicted_name]

    }
```
